```python
highs = weather_data.loc[:,["high"]].to_numpy()
lows = weather_data.loc[:,["low"]].to_numpy()
avgs = weather_data.loc[:,["avg"]].to_numpy()
```

```python
def min_max_avg_annual_temp(mins, maxs) -> np.float64:
    min_sum = np.sum(mins)
    print(min_sum)
    max_sum = np.sum(maxs)
    print(max_sum)
    sample_size = mins.size + maxs.size
    print(sample_size)
    avg_annual_temp = (min_sum + max_sum) / sample_size
    return avg_annual_temp
```

```python
min_max_avg_annual_temp(lows, highs)
```
