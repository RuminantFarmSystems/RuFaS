```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output/CSVs/om/saved_variables_csv_eric_alfalfa_2014_filter.txt_10-Oct-2023_Tue_09-13-48.csv")
```

```python
data.columns
```

```python
days = data["Time.day.values"].tolist()
years = data["Time.year.values"].tolist()
dates = [f"{day}/{year}" for day, year in zip(days, years) if not (day < 106 and year == 2014) and  not (246 < day and year == 2016)]
```

```python
total_biomass = data["field='field',crop='default alfalfa',planted=105,2014.biomass(kg/ha).values"].dropna().tolist()
biomass_increase = data["field='field',crop='default alfalfa',planted=105,2014.biomass_growth(kg/ha).values"].dropna().tolist()
biomass_increase.insert(0, 0)
surface_biomass = data["field='field',crop='default alfalfa',planted=105,2014.above_ground_biomass(kg/ha).values"].dropna().tolist()
root_biomass = data["field='field',crop='default alfalfa',planted=105,2014.root_biomass(kg/ha).values"].dropna().tolist()
root_biomass.insert(0, 0)
```

```python
zipped_surface_root_biomasses = list(zip(surface_biomass, root_biomass))
surface_root_biomasses = [mass[0] + mass[1] for mass in zipped_surface_root_biomasses]
```

```python
x = dates
x_ticks = x[::100]
plt.figure(figsize=(15, 4))
plt.plot(x, total_biomass, "o--", alpha=0.5)
plt.plot(x, biomass_increase, "o-", alpha=0.5)
plt.xticks(x_ticks)
plt.xlabel("Time (day/year)")
plt.ylabel("Biomass (kg/ha)")
plt.legend(["Total biomass", "Biomass change"])
plt.show()
```

```python
x = dates
x_ticks = x[::10]
plt.figure(figsize=(15, 4))
plt.plot(x, total_biomass, "o-", alpha=0.5)
plt.plot(x, surface_biomass, "o--", alpha=0.5)
plt.plot(x, root_biomass, "o:", alpha=0.5)
plt.xlabel("Time (day/year)")
plt.ylabel("Biomass (kg/ha)")
plt.legend(["Total Biomass", "Surface Biomass", "Root Biomass"])
plt.xticks(x_ticks)
plt.show()
```

```python
x = dates
x_ticks = x[::10]
plt.figure(figsize=(15, 4))
plt.plot(x, total_biomass, "o-", alpha=0.5)
plt.plot(x, surface_root_biomasses, "o-.", alpha=0.5)
plt.xticks(x_ticks)
plt.xlabel("Time (day/year)")
plt.ylabel("Biomass (kg/ha)")
plt.legend(["Total Biomass", "Surface + Root Biomass"])
plt.show()
```
